# Visualize
* Learn to use matplotlib
* subplots
* multiple plots
* bar plots

```python
data=pd.read_csv("SOL-USD.csv",index_col='Date',parse_dates=True)
```

```python
data
```

```python
data.plot()
```

```python
data['Close'].plot()
```

```python
fig,ax=plt.subplots()
data['Close'].plot(ax=ax)
ax.set_ylabel("Price")
ax.set_title("Solana/USD")
```

```python
fig,ax=plt.subplots()
data['Open'].plot(ax=ax)
ax.set_ylabel('Price')
ax.set_title("Solan/USD open price")
```

```python
fig,ax=plt.subplots(2,2)
data['Open'].plot(ax=ax[0,0],title="Open")
data['High'].plot(ax=ax[0,1],title="Close")
data['Low'].plot(ax=ax[1,0],title='Low')
data['Close'].plot(ax=ax[1,1],title='Close')
plt.tight_layout()
```
