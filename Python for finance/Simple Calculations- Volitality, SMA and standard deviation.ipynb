# Standard Deviation
* standard deviation(p)= standard deviation (daily)*square root of (p)
*standard deviation(anually)=standard deviation (daily)*squareroot (252)

```python
data=pd.read_csv("SOL-USD.csv",index_col="Date",parse_dates=True)
```

```python
data['per_chg']=data['Close'].pct_change()
```

```python
data
```

```python
data['Log returns']=np.log(data['Close']/data['Close'].shift())
```

```python
data
```

```python
data['Close']/data['Close'].shift()
```

```python
data['Close'].shift()
```

```python
daily_dev=data['Log returns'].std()
```

```python
daily_dev
```

```python
voltality=daily_dev*(252**.5)
```

```python
voltality
```

```python
str_vol=str(round(voltality,3)*100)
```

```python
str_vol
```

```python
fig,ax=plt.subplots()
data['Log returns'].hist(ax=ax,color='purple')
ax.set_xlabel("log returns")
ax.set_ylabel("freq of log return")
ax.set_title("SOLUSD volatility "+str_vol)
```
